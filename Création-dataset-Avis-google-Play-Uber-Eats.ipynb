```python
from google_play_scraper import reviews, Sort
import pandas as pd

# R√©cup√©rer 200 avis Uber Eats France
result, _ = reviews(
    'com.ubercab.eats',
    lang='fr',
    country='FR',
    sort=Sort.NEWEST,
    count=200
)

# Transformer en DataFrame
df = pd.DataFrame(result)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
df.head()
```

```python
df = df.drop(columns=['userImage'])
df = df.drop(columns=['userName'])
df = df.drop(columns=['thumbsUpCount'])
df = df.drop(columns=['reviewCreatedVersion'])
df = df.drop(columns=['replyContent'])
df = df.drop(columns=['repliedAt'])
df = df.drop(columns=['appVersion'])
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
df.head()
```
